```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
# кол-во вихрей
N = 4
# интенсивность вихрей
G = np.ones(N)
# координаты вихрей
Cord = np.zeros((N,2))
Cord = [[1,1], [1,-1], [-1, -1], [-1, 1]]
#
tay = 0.1
# стартовая точка
XY0 = [1.1, 1.1]
#
t, T = 0, 100*math.pi
```

```python
def NewDot (XYn):
    XYn1 = [0,0]
    for i in range(N):
        s = ((XYn[0]- Cord[i][0])**2 + (XYn[1]- Cord[i][1])**2)*2*math.pi
        XYn1[0] += G[i]*(-XYn[1] + Cord[i][1])/s
        XYn1[1] += G[i]*(XYn[0]- Cord[i][0])/s
    XYn1[0] = XYn[0] + tay*XYn1[0]
    XYn1[1] = XYn[1] + tay*XYn1[1]
    return XYn1
def PrintPlot (XY0 = XY0, t = t, T = T):
    fig = plt.figure()
    ax = fig.add_axes([-2,-2,2,2])
    for i in range(N):
        plt.scatter(Cord[i][0], Cord[i][1], s = 30 , c = 'r', marker = 'x')
    tn = t
    XYn = XY0
    #Выделим первую точку
    plt.scatter(XYn[0], XYn[1], s = 20, c = 'r')
    while (tn <= T):
        XYn = NewDot(XYn)
        plt.scatter(XYn[0], XYn[1], s = 1, c = 'b')
        tn += tay
    #Выделим послежнюю точку
    plt.scatter(XYn[0], XYn[1], s = 30, c = 'y')
    plt.show()
```

```python
PrintPlot()
```

```python
PrintPlot([0.5,0.5])
```

```python
PrintPlot([0.1, 0.1])
```
